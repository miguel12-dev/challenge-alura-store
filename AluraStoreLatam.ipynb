### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_total_tienda1 = tienda['Precio'].sum()
print(f"Ingreso total para Tienda 1: {ingreso_total_tienda1:,.2f}")
```

```python
ingreso_total_tienda2 = tienda2['Precio'].sum()
print(f"Ingreso total para Tienda 2: {ingreso_total_tienda2:,.2f}")
```

```python
ingreso_total_tienda3 = tienda3['Precio'].sum()
print(f"Ingreso total para Tienda 3: {ingreso_total_tienda3:,.2f}")
```

```python
ingreso_total_tienda4 = tienda4['Precio'].sum()
print(f"Ingreso total para Tienda 4: {ingreso_total_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria_tienda = tienda.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Ventas por categoría en Tienda 1:\n", ventas_por_categoria_tienda.head())
```

```python
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("Ventas por categoría en Tienda 2:\n", ventas_por_categoria_tienda2.head())

ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("\nVentas por categoría en Tienda 3:\n", ventas_por_categoria_tienda3.head())

ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
print("\nVentas por categoría en Tienda 4:\n", ventas_por_categoria_tienda4.head())
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
print(f"Calificación promedio para Tienda 1: {calificacion_promedio_tienda1:.2f}")

calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
print(f"Calificación promedio para Tienda 2: {calificacion_promedio_tienda2:.2f}")
```

```python
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
print(f"Calificación promedio para Tienda 3: {calificacion_promedio_tienda3:.2f}")

calificacion_promedio_tienda4 = tienda4['Calificación'].mean()
print(f"Calificación promedio para Tienda 4: {calificacion_promedio_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
print('--- Tienda 1 ---')
productos_vendidos_tienda1 = tienda['Producto'].value_counts()
print("Productos mas vendidos:\n", productos_vendidos_tienda1.head())
print("\nProductos menos vendidos:\n", productos_vendidos_tienda1.tail())
```

```python
print('\n--- Tienda 2 ---')
productos_vendidos_tienda2 = tienda2['Producto'].value_counts()
print("Productos mas vendidos:\n", productos_vendidos_tienda2.head())
print("\nProductos menos vendidos:\n", productos_vendidos_tienda2.tail())

```

```python
print('\n--- Tienda 3 ---')
productos_vendidos_tienda3 = tienda3['Producto'].value_counts()
print("Productos mas vendidos:\n", productos_vendidos_tienda3.head())
print("\nProductos menos vendidos:\n", productos_vendidos_tienda3.tail())

```

```python
print('\n--- Tienda 4 ---')
productos_vendidos_tienda4 = tienda4['Producto'].value_counts()
print("Productos mas vendidos:\n", productos_vendidos_tienda4.head())
print("\nProductos menos vendidos:\n", productos_vendidos_tienda4.tail())
```

# 5. Envío promedio por tienda

```python
promediio_envio_tienda1 = tienda['Costo de envío'].mean()
print(f"Costo de envio promedio para Tienda 1: {promediio_envio_tienda1:.2f}")

promediio_envio_tienda2 = tienda2['Costo de envío'].mean()
print(f"Costo de envio promediio para Tienda 2: {promediio_envio_tienda2:.2f}")

promedioo_envio_tienda3 = tienda3['Costo de envío'].mean()
print(f"Costo de envio promedio para Tienda 3: {promedioo_envio_tienda3:.2f}")

promediio_envio_tienda4 = tienda4['Costo de envío'].mean()
print(f"Costo de envio promedioo para Tienda 4: {promediio_envio_tienda4:.2f}")
```

# **6. Graficos**

```python
import matplotlib.pyplot as plt

# Primer grafico: Ingreso total por tienda (en  un grafico de barras)

plt.figure(figsize=(10, 6))

ingresos = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
tiendas_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(tiendas_nombres, ingresos, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Nombre de la Tienda') # etiqeta para el eje x
plt.ylabel('Ingreso Total (Millones)') # etiqeta para el eje y
plt.title('Ingreso Total para Cada Tienda (Grafico de Barra)') # titulo del grafico
plt.grid(axis='y', linestyle='--') # para que se vea mejor
plt.tight_layout()
plt.show() #Impresion del grafico
```

```python

# Segundo grafico: Ventas por categoria en Tienda 1 (un grafico comoo un pastel)

plt.figure(figsize=(10, 8))
ventas_por_categoria_tienda.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.ylabel('') # no nesesitamos etiqeta en el eje y para un pastel
plt.title('Porcentaje de Ventas por Categoria en Tienda 1') # otro titulo
plt.tight_layout()
plt.show()

```

```python
# Tercer grafico: Calificacion promediio de clientes por tienda (otro grafico de barras para mejor explicacion)

plt.figure(figsize=(10, 6))

calificaciones = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]
plt.bar(tiendas_nombres, calificaciones, color=['gold', 'orange', 'brown', 'pink'])
plt.xlabel('Tienda')
plt.ylabel('Calificacion Promediio')
plt.title('Calificacion Promediio de Clientes por Tienda')
plt.ylim(0, 5) # las calificaciones van de 0 a 5
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```
